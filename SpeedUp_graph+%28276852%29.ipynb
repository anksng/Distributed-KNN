```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
time_one_process=20.150286197662354 
Num_workers = [1,2,4,6,8,10]

time_with_mpi=np.array([20.150286197662354,20.049756002426147,19.0701818466187,23.875270128250122,26.06355166435242,28.198495388031006])

speed_up = []
for i in range(6):
    s=  time_one_process/time_with_mpi[i]
    speed_up.append(s)
print("Speed up values are", speed_up)

plt.plot(Num_workers,speed_up,'b',label="speed up line")

plt.xlabel('Num of processes')
plt.ylabel('speedup')
plt.title("Speedup Graph")
plt.legend()
plt.show()
```

```python
plt.plot(Num_workers,time_with_mpi,'r',label="speed up line")

plt.xlabel('Num of processes')
plt.ylabel('time')
plt.title("Time vs num of processes")
plt.legend()
plt.show()
```
